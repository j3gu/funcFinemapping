### Summary
By comparing eQTL and GWAS discovery, the author found they differ in the types of variants and genes. By modeling under selections, they were able to observe how limited overlaps of eQTLs and GWAS hits are expected.

### Background

*General challenge*:   
Uncover the functional mechanisms of GWAS hits, among which ~90% located in the non-coding regions. GWAS signals are mainly located in open chromatin regions and enriched with regulatory elements and eQTLs. This suggests most GWAS hits affect traits through altering gene regulation of nearby genes [ref 2-7]. 

*Specific challenge*:  
most GWAS hits are not explained by known eQTLs called in various cell types.
* a median of 21% GWAS hits per traits colocalize with eQTLs. 
* only 11% of heritability or complex traits mediated by gene expression in GTEx tissues [ref 18].
    * Haky - This underestimation needs to be updated with current version of GTEx. 

*Explanations*:
1. Current GTEx is lack of power to identify eQTLs for rare cell types or specific contexts. An eQTL study on multiple stages of iPSCs contributes to ~10% more colocalizations with nuerological traits [ref 25]. However, it is unknown how much of the cap can be resolved. 
2. Incomplete statistical power for the colocalization methods even if mapping performed on the correct contexts [ref 18,32].
3. Other mechanisms: post-transcriptional regulation

*Proposals*:
- Compare the discovery between eQTLs from 38 tissues and GWAS hits from 44 complex traits from three aspects:
    * selective constraints: pLI scores (predicted loss-of-function scores)
    * transcription regulatory activity: enhancer activity and TSS usage
- Build a model of variant discovery to explain the observations

### Key results
The author found the differences in the discovered variants and genes between GWAS and eQTL studies. Each GWAS hit or each eQTL is the lead SNP for the association. To make it more comparable with the same reference, the lead SNPs were assigned to the nearest genes for both methods. 

**_Challenge_**:  
The differences between GWAS and eQTL discovery may be confounded by systematic differences like MAF and gene density. A control set of variants with matched MAF and density were run to correct for the confounding effects.

Compared with the matched ones, eQTL genes show overall depletion of high-pLI genes.

### Method - Model the discovery outcomes under different selection models

**_Model discovery regions_**   
Parameter spaces of $\beta^2$ and $\gamma^2$ were computed using the expected contribution of variants to phenotypic variance.
$$
E[Var(G)\beta^2\gamma^2|\beta, \gamma] \propto \beta^2\gamma^2
$$

**_Discovery thresholds vs. sample size_**  
Let's considre a simple linear regression model that estimates the effect of a single SNP, $\beta$, on a quantitative phenotype Y. 
$$
Y = G\beta + \epsilon
$$

For a given SNP, its chi-square statistics $\chi^2$ needs to be greater than a significance threshold for it to be discovered. 
$$
\chi^2 = \frac{\hat \beta^2}{Var(\hat \beta)} \approx \frac{n\hat \beta^2 Var(G)}{Var(Y)} > \chi^2_c,
$$
when we assume the effect of individual SNP is negligible. 
Thus,$c^*$ is the threshold that discovered variants are supposed to achieve the following condition:
$$
\beta^2Var(G) > c^*
$$

This condition is in expectation satisfied if 
$$
\begin{align}
\beta^2 Var(G) &= E[\hat \beta^2]Var(G) \\
             &> \frac{\chi^2_c Var(Y)}{n} \\
\end{align}
$$
